Aqui eu cheguei na quantidade de alunos por genero.

```python
media_por_sexo = df.groupby("sex")["G3"].mean()
print(media_por_sexo)

media_por_sexo = df.groupby("sex")["G3"].mean().round(2)
print(media_por_sexo)
```

```python
import seaborn as sns
coluna_idade = df['age'][0]
coluna_nota_final = df['G3'][0]
print(coluna_idade)
print(coluna_nota_final)
# Agrupar média de G3 por sexo
media_por_sexo = df.groupby("sex")["G3"].mean().round(2)

# Estilo dos gráficos
sns.set(style="whitegrid")

# 1. Gráfico de Barras
plt.figure(figsize=(6, 4))
sns.barplot(x=media_por_sexo.index, y=media_por_sexo.values, palette="pastel")
plt.title("Média da Nota Final (G3) por Sexo")
plt.xlabel("Sexo")
plt.ylabel("Média da Nota Final (G3)")
plt.tight_layout()
plt.show()

# 2. Gráfico de Linha: G3 por idade, separado por sexo
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x="age", y="G3", hue="sex", marker="o", palette="Set2")
plt.title("Nota Final (G3) por Idade e Sexo")
plt.xlabel("Idade")
plt.ylabel("Nota Final (G3)")
plt.tight_layout()
plt.show()

# 3. Gráfico de Dispersão: Faltas x Nota Final
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="absences", y="G3", hue="sex", palette="cool", alpha=0.7)
plt.title("Relação entre Faltas e Nota Final (G3)")
plt.xlabel("Faltas")
plt.ylabel("Nota Final (G3)")
plt.tight_layout()
plt.show()
```

No primeiro grafico em Barras, ficou claro que o genero feminino tiveram o melhor desempenho em relação ao genero masculino. No segundo de Linhas, visualiza-se por idade de cada genero. E no ultimo revelou como as faltas de cada genero influenciou nas notas.
